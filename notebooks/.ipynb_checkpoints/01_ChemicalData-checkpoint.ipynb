# Chemical Data: Structured Data

```python
import pandas as pd
```

```python
X = pd.read_csv('../data/molorg_features.csv',header = 0)
print X.shape
X.head(n=10)
```

```python
y = pd.read_csv('../data/molorg_pol.csv', header = 0)
print y.shape
y.head(n=10)
```

# check for missing data

```python
print X.isna().any().sum()
print y.isna().any().sum()
# available options to replace or remove missing data
# pandas: fillna/dropna
# sklearn: imputer
```

# check for constant columns

```python
pre_columns = X.columns
X = X.loc[:, (X != X.iloc[0]).any()]
print X.shape
print 'removed columns: ', [i for i in pre_columns if i not in X.columns]
```

```python
Xy = pd.concat([X,y], axis = 1)
print Xy.shape
```

```python
Xy_corr = Xy.corr(method='pearson')
```

```python
Xy_corr[y.columns[0]].sort_values(ascending=False).head(n = 10)
```

```python
Xy_corr[y.columns[0]].sort_values(ascending=False).tail(n = 10)
```
